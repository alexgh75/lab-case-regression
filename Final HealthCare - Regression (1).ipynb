### 3. Split the data into a training set and a test set.

```python
X = all_data
y = target['TARGET_B']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

### 4. Split further into train_num and train_cat. Also test_num and test_cat.¶

```python
X_train_num = X_train.select_dtypes('number')
X_train_cat = X_train.select_dtypes(object)

X_test_num = X_test.select_dtypes('number')
X_test_cat = X_test.select_dtypes(object)
```

```python
print(X_train_num.shape)
print(X_train_cat.shape)
print(X_test_num.shape)
print(X_test_cat.shape)
```

```python
print(y_train.shape)
print(y_test.shape)
```

### 5. Scale the features either by using MinMax Scaler (train_num, test_num)¶

```python
from sklearn.preprocessing import MinMaxScaler

transformer = MinMaxScaler().fit(X_train_num)
```

```python
# Scaling
X_train_scaled = pd.DataFrame(transformer.transform(X_train_num), columns=X_train_num.columns)
X_test_scaled = pd.DataFrame(transformer.transform(X_test_num), columns=X_test_num.columns)
```

### 6. Encode the categorical features using One-Hot Encoding (train_cat, test_cat)

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
X_train_cat.head()
```

```python
ordinal_list = ['STATE','HOMEOWNR', 'GENDER', 'RFA_2R', 'RFA_2A', 'GEOCODE2', 'DOMAIN_A']
```

```python
# Fitting the encoder
encoder = OneHotEncoder(drop='first', handle_unknown = 'ignore').fit(X_train_cat)
column_name = encoder.get_feature_names_out(X_train_cat.columns)
```

### 7. Fit only on train data, transform both train and test

```python
# Encoding
X_train_cat = pd.DataFrame(encoder.transform(X_train_cat).toarray(), columns = column_name)
X_test_cat = pd.DataFrame(encoder.transform(X_test_cat).toarray(), columns = column_name)
```

```python
print(X_train_cat.shape)
```

### 8. Re-concatenate train_num and train_cat as X_train as well as test_num and test_cat as X_test

```python
X_train_processed = pd.concat([X_train_scaled,X_train_cat], axis = 1)
X_test_processed = pd.concat([X_test_scaled,X_test_cat], axis = 1)
```

```python
from sklearn.linear_model import LogisticRegression

classification = LogisticRegression(random_state =0, solver='lbfgs').fit(X_train_processed,y_train)
classification.score(X_test_processed, y_test)
```

### Managing imbalance in the dataset

### 1. Check for the imbalance.
### 2. Use the resampling strategies used in class for upsampling and downsampling to create a balance between the two classes.
### 3. Each time fit the model and see how the accuracy of the model has changed.

```python
from sklearn.utils import resample

y_train = y_train.reset_index(drop=True)
X_train = X_train_processed.reset_index(drop=True)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
train_data = pd.concat([X_train, y_train], axis=1)
print(train_data.shape)
train_data
```

### Down Sample

```python
category_0 = train_data[train_data['TARGET_B'] == 0]
category_1 = train_data[train_data['TARGET_B'] == 1]



category_0_downsampled = resample(category_0,
                                  replace=False,
                                  n_samples=len(category_1),
                                  random_state=42)


resampled_down = pd.concat([category_0_downsampled, category_1], axis=0)
```

```python
resampled_down
```

```python
y_downsampled = resampled_down['TARGET_B']
X_downsampled = resampled_down.drop(columns=['TARGET_B'])
```

```python
classification_downsampled = LogisticRegression(random_state =0, solver='lbfgs').fit(X_downsampled,y_downsampled)
print(classification_downsampled.score(X_downsampled, y_downsampled))
```

```python
predictions_downsampled = classification_downsampled.predict(X_test_processed)
confusion_matrix(y_test,predictions_downsampled)
#predicted 1's but still has 7387 0 which should have been 1 and 428 1's which should have been 0
```

### Up Sampled

```python
#upsampling
category_1_oversampled = resample(category_1,
                                  replace=True,
                                  n_samples = len(category_0))
```

```python
data_upsampled = pd.concat([category_0, category_1_oversampled], axis=0)
data_upsampled
```

```python
y_upsampled = data_upsampled['TARGET_B']
X_upsampled = data_upsampled.drop(columns=['TARGET_B'])
```

```python
classification_upsampled = LogisticRegression(random_state =0, solver='lbfgs').fit(X_upsampled,y_upsampled)
print(classification_upsampled.score(X_upsampled, y_upsampled))
```

```python
predictions_upsampled=classification_upsampled.predict(X_test_processed)
confusion_matrix(y_test,predictions_upsampled)
#predicted 1's but still has 6930 0 which should have been 1 and 450 1's which should have been 0
```

### Final regression model in "Health Care for All" Case

```python
all_donation = pd.concat((numerical, categorical, target),axis=1)
print(all_donation.shape)
all_donation.head()
```

```python
donations_data = all_donation[all_donation['TARGET_B']==1]
print(donations_data.shape)
donations_data.head()
```

```python
X_don = donations_data.drop(columns=['TARGET_B','TARGET_D'])
y_don = donations_data['TARGET_D']
```

```python
print(X_don.shape)
X_don.head()
```

```python
print(y_don.shape)
y_don.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_don, y_don, test_size=0.2, random_state=0)

numericalX_train   = X_train.select_dtypes(np.number)
numericalX_test    = X_test.select_dtypes(np.number)
categoricalX_train = X_train.select_dtypes(object)
categoricalX_test  = X_test.select_dtypes(object)

# OneHotEncode the categoricals

from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(drop='first').fit(categoricalX_train)
encoded_categorical_train = encoder.transform(categoricalX_train).toarray()
encoded_categorical_train = pd.DataFrame(encoded_categorical_train, columns=encoder.get_feature_names_out()) # needed to avoid error
encoded_categorical_test = encoder.transform(categoricalX_test).toarray()
encoded_categorical_test = pd.DataFrame(encoded_categorical_test, columns=encoder.get_feature_names_out())

# Scale the numericals

from sklearn.preprocessing import MinMaxScaler
transformer = MinMaxScaler().fit(numericalX_train)
scaled_numerical_train = transformer.transform(numericalX_train)
scaled_numerical_train = pd.DataFrame(scaled_numerical_train, columns=numericalX_train.columns)
scaled_numerical_test  = transformer.transform(numericalX_test)
scaled_numerical_test  = pd.DataFrame(scaled_numerical_test, columns=numericalX_train.columns)

X_train_treated = pd.concat([scaled_numerical_train, encoded_categorical_train], axis = 1)
X_test_treated  = pd.concat([scaled_numerical_test,  encoded_categorical_test],  axis = 1)
```

```python
from sklearn.linear_model import LinearRegression as LinReg

linreg = LinReg()    # model
linreg.fit(X_train_treated, y_train)   # model training
y_pred_linreg = linreg.predict(X_test_treated)   # model prediction
```

```python
print ('train R2: {} -- test R2: {}'.format(linreg.score(X_train_treated, y_train),
                                            linreg.score(X_test_treated, y_test)))
```

```python
from sklearn.metrics import mean_squared_error as mse


train_mse = mse(linreg.predict(X_train_treated), y_train)
test_mse = mse(y_pred_linreg, y_test)

print ('train MSE: {} -- test MSE: {}'.format(train_mse, test_mse))
```

```python
print ('train RMSE: {} -- test RMSE: {}'.format(train_mse**.5, test_mse**.5))
```

```python
from sklearn.metrics import mean_absolute_error as mae

train_mae = mae(linreg.predict(X_train_treated), y_train)
test_mae = mae(linreg.predict(X_test_treated), y_test)

print ('train MAE: {} -- test MAE: {}'.format(train_mse, test_mse))
```

```python
from sklearn.tree import DecisionTreeRegressor
model1 = DecisionTreeRegressor()
from sklearn.linear_model import LinearRegression
model2 = LinearRegression()
from sklearn.neighbors import KNeighborsRegressor
model3 = KNeighborsRegressor()

from sklearn.model_selection import cross_val_score
model_pipeline = [model1, model2, model3]
model_names = ['Decision Tree Regressor', 'Linear Regression', 'KNN']
scores = {}
for model, model_name in zip(model_pipeline, model_names):
    mean_score = np.mean(cross_val_score(model, X_train_treated, y_train, cv=5))
    scores[model_name] = mean_score
print(scores)
```
